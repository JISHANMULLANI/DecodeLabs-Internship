```python
import pandas as pd
```

```python
df = pd.read_excel("Dataset for Data Analytics.xlsx")
```

```python
df.head()
```

```python
df.tail()
```

```python
# Missing values
print("Missing Values:")
df.isnull().sum()
```

```python
# fill duplicate value with No Coupon
df["CouponCode"].fillna("No Coupon", inplace=True)
```

```python
df["Date"]
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Date"]
```

```python
# remove extra space
df["Product"] = df["Product"].str.strip()
```

```python
# remove extra space
df["OrderStatus"] = df["OrderStatus"].str.strip()
```

```python
df.to_excel("Cleaned_Dataset.xlsx", index=False)
```
