```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Dataset
df = pd.read_excel("Dataset for Data Analytics.xlsx")
```

```python
df.head()
```

```python
# shape of dataset
print("Dataset Shape:", df.shape)
```

```python
# Basic Statistics
print("\nBasic Statistics:")
print(df.describe())
```

```python
# Mean, Median and Count
print("\nQuantity")
print("Mean:", df["Quantity"].mean())
print("Median:", df["Quantity"].median())
print("Count:", df["Quantity"].count())
```

```python
print("\nTotal Price")
print("Mean:", df["TotalPrice"].mean())
print("Median:", df["TotalPrice"].median())
print("Count:", df["TotalPrice"].count())
```

```python
# Product Distribution
print("\nProduct Counts:")
print(df["Product"].value_counts())
```

```python
# Payment Method Distribution
print("\nPayment Methods:")
print(df["PaymentMethod"].value_counts())
```

```python
# Order Status Distribution
print("\nOrder Status:")
print(df["OrderStatus"].value_counts())
```

```python
# Outlier Detection using IQR
Q1 = df["TotalPrice"].quantile(0.25)
Q3 = df["TotalPrice"].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
outliers = df[
    (df["TotalPrice"] < (Q1 - 1.5 * IQR)) |
    (df["TotalPrice"] > (Q3 + 1.5 * IQR))
]
```

```python
print("\n Number of Outliers:", len(outliers))
```

```python
# Create Box Plot
plt.figure(figsize=(8,5))
plt.boxplot(df["TotalPrice"])

plt.title("Box Plot of TotalPrice")
plt.ylabel("Total Price")

plt.show()
```
